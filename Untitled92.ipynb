```python
# ===== 1️⃣ Install dependencies =====
# pip install tensorflow opencv-python matplotlib pandas

import cv2
import numpy as np
import tensorflow as tf
from tensorflow.keras.applications.mobilenet_v2 import MobileNetV2, preprocess_input, decode_predictions
from tensorflow.keras.preprocessing import image
import pandas as pd
import datetime
import sys

# ===== 2️⃣ Load Pre-trained Model =====
print("Loading MobileNetV2 model...")
model = MobileNetV2(weights='imagenet')
print("Model loaded!")

# ===== 3️⃣ Robust Webcam Detection =====
def get_working_camera(max_indices=5):
    for i in range(max_indices):
        try:
            cap = cv2.VideoCapture(i)
            if cap is not None and cap.isOpened():
                print(f"Using camera index {i}")
                return cap
            else:
                cap.release()
        except:
            continue
    return None

cap = get_working_camera(10)
if cap is None:
    print("Error: Could not open any webcam. Make sure no other app is using it and drivers are installed.")
    sys.exit(1)

# ===== 4️⃣ Prepare CSV log =====
log_file = "predictions_log.csv"
df = pd.DataFrame(columns=["Timestamp", "Prediction", "Confidence"])
df.to_csv(log_file, index=False)

# ===== 5️⃣ Real-Time Recognition Loop =====
print("Press 'q' to quit the demo.")

while True:
    if cap is None or not cap.isOpened():
        print("Webcam disconnected or failed.")
        break

    ret, frame = cap.read()
    if not ret or frame is None:
        print("Failed to grab frame. Skipping...")
        continue

    # Preprocess frame
    img = cv2.resize(frame, (224,224))
    img_array = image.img_to_array(img)
    img_array = np.expand_dims(img_array, axis=0)
    img_array = preprocess_input(img_array)

    # Prediction
    preds = model.predict(img_array)
    decoded = decode_predictions(preds, top=1)[0][0]
    label = f"{decoded[1]}: {decoded[2]*100:.2f}%"

    # Overlay prediction
    cv2.putText(frame, label, (10,30), cv2.FONT_HERSHEY_SIMPLEX,
                1, (0,255,0), 2, cv2.LINE_AA)
    cv2.imshow("AI Image Recognition Demo", frame)

    # Log prediction
    timestamp = datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S")
    df = pd.DataFrame([[timestamp, decoded[1], float(decoded[2])]], columns=["Timestamp","Prediction","Confidence"])
    df.to_csv(log_file, mode='a', header=False, index=False)

    # Quit on 'q'
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# ===== Safe Release =====
if cap is not None and cap.isOpened():
    cap.release()
cv2.destroyAllWindows()
print(f"Demo ended. Predictions logged to {log_file}")
```
