```python
import cv2
import matplotlib.pyplot as plt
image_path='example.jpg'
image=cv2.imread(image_path)
image_rgb=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
height,width, _=image_rgb.shape
rectl_width, rectl_height=150,150
top_left1=(20,20)
bottom_right1=(rectl_width+20,rectl_height+20)
cv2.rectangle(image_rgb,top_left1,bottom_right1,(0,255,0),2)
```

```python
rect2_width,rect2_height=200,150
top_left2=(width-rect2_width-20,height-rect2_height-20)
bottom_right2=(width-20,height-20)
cv2.rectangle(image_rgb,top_left2,bottom_right2,(0,255,0),2)
```

```python
center1_x=top_left1[0]+rectl_width//2
center1_x=top_left1[1]+rectl_height//2
center2_x=top_left2[0]+rect2_width//2
center2_y=top_left2[1]+rect2_height//2
cv2.circle(image_rgb,(center1_x,center1_x),15,(0,0,255),-1)
cv2.circle(image_rgb,(center2_x,center2_y),15,(0,255,0),-1)
```

```python
cv2.line(image_rgb,(center1_x,center1_x),(center2_x,center2_y),(255,0,0),2)
```

```python
font=cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(image_rgb, 'region 1', (top_left1[0], top_left1[1]-10), font, 0.5, (255, 255, 255), 1)
cv2.putText(image_rgb, 'region 2', (top_left2[0], top_left2[1]-10), font, 0.5, (255, 255, 255), 1)
cv2.putText(image_rgb, 'center 1', (center1_x, center1_x-10), font, 0.5, (255, 255, 255), 1)
cv2.putText(image_rgb, 'center 2', (center2_x, center2_y-10), font, 0.5, (255, 255, 255), 1)
```

```python
arrow_start=(width-50,20)
arrow_end=(width-50,height-20)
cv2.arrowedLine(image_rgb,arrow_start,arrow_end,(0,0,255),3, tiplength=0.7)
cv2.arrowedLine(image_rgb,arrow_end,arrow_start,(0,0,255),3, tiplength=0.7)
```

```python
height_label_position=(arrow_start[0]-150, (arrow_start[1]+arrow_end[1])//2)
cv2.putText(image_rgb, 'height', height_label_position, font, 0.5, (255, 255, 255), 1, cv2.LINE_AA)
```

```python
plt.figure(figsize=(10,10))
plt.imshow(image_rgb)
plt.title('image_rgb')
plt.axis('off')
plt.show()
```
