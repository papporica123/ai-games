```python
import requests
from google.colab import userdata

api_url="https://api-inference.huggingface.co/models/distilbert-base-uncased"

# Securely fetch your Hugging Face token from Colab secrets
# You need to add your HF token to Colab secrets under the name 'HF_TOKEN'
hf_token = userdata.get('HF_TOKEN')

header={"Authorization":f"Bearer {hf_token}"}
text="I love this movie! It was fantastic."

response=requests.post(api_url,headers=header,json={"inputs":text})

if response.status_code==200:
  result=response.json()
  print(f"sentiment:{result[0]['label']} with confidence score:{result[0]['score']}")
else:
    print(f"error:{response.status_code}")
```
